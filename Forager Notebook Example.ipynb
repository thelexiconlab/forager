```python
import forager
```

Create a small sample data file.

```python
sample_data = '''
23	cat
23	dog
23	horse
12	fish
12	sea_turtle
12	octopus
'''
with open('sample.txt', 'w') as sample_file:
  sample_file.write(sample_data)
```

Use forager to retreive the data file for processing

```python
data = forager.retrieve_data('sample.txt')
```

And finally, "run" the model

```python
forager.run_model(data, 'dynamic', 'all', 'sample')
```
